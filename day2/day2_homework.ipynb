```python
# ===== English cell English =====
random.seed(int(hashlib.md5(STUDENT_ID.encode()).hexdigest()[:8], 16))

all_topics = [
    {'topic': 'healthcare', 'text': 'English English AI English X-ray English English 30 English 5 English English Deep Learning English 95% English'},
    {'topic': 'banking', 'text': 'English English KADE AI Assistant English RAG English English 5 English 30 English English'},
    {'topic': 'education', 'text': 'English English RAG English-English English PDF English 500 English vector embeddings English multilingual model English'},
    {'topic': 'agriculture', 'text': 'English English AI English English English 92% English 40%'},
    {'topic': 'logistics', 'text': 'Kerry Express English AI English English 15% English 200 English English Machine Learning English'},
    {'topic': 'retail', 'text': 'English English English AI Recommendation System English English 25% English 10 English'},
    {'topic': 'energy', 'text': 'English (EGAT) English AI English English 8% English Deep Learning English Smart Meter English'},
    {'topic': 'tourism', 'text': 'English. English AI Chatbot English English 5 English English 1 English English call center English 60%'},
    {'topic': 'insurance', 'text': 'English English AI English English English 7 English 1 English English fraud English 90%'},
    {'topic': 'manufacturing', 'text': 'SCG English AI English Computer Vision English English 30% Predictive Maintenance English 2 English'},
]

random.shuffle(all_topics)
my_topics = all_topics[:6]  # English 6 English 10 English

# English Qdrant
embed_model = SentenceTransformer('intfloat/multilingual-e5-large')
splitter = RecursiveCharacterTextSplitter(chunk_size=150, chunk_overlap=30)

my_chunks = []
my_sources = []
for item in my_topics:
    chunks = splitter.split_text(item['text'])
    for c in chunks:
        my_chunks.append(c)
        my_sources.append(item['topic'])

# Embed + Index
passages = [f'passage: {c}' for c in my_chunks]
embeddings = embed_model.encode(passages)
VECTOR_SIZE = embeddings.shape[1]

qdrant = QdrantClient(':memory:')
collection_name = f'hw2_{STUDENT_ID}'
qdrant.create_collection(
    collection_name=collection_name,
    vectors_config=models.VectorParams(size=VECTOR_SIZE, distance=models.Distance.COSINE)
)

points = [
    models.PointStruct(id=i, vector=embeddings[i].tolist(),
                       payload={'text': my_chunks[i], 'source': my_sources[i]})
    for i in range(len(my_chunks))
]
qdrant.upsert(collection_name=collection_name, points=points)

# English query English
query_pool = [
    'AI English',
    'English',
    'English Machine Learning',
    'Chatbot English',
    'Computer Vision English',
    'English Deep Learning',
]
random.shuffle(query_pool)
MY_QUERY = query_pool[0]

# English tool type
tool_types = [
    {'name': 'English ROI English AI', 'func': 'calculate_ai_roi', 'params': 'investment_baht: float, savings_per_year: float, years: int'},
    {'name': 'English AI', 'func': 'assess_ai_readiness', 'params': 'num_employees: int, has_data: bool, budget_million: float'},
    {'name': 'English Accuracy Score', 'func': 'calculate_accuracy', 'params': 'true_positives: int, false_positives: int, false_negatives: int, true_negatives: int'},
    {'name': 'English AI Model', 'func': 'recommend_model', 'params': 'task_type: str, data_size: str, budget: str'},
]
random.shuffle(tool_types)
MY_TOOL = tool_types[0]

# Workflow type
workflow_types = ['SequentialAgent', 'ParallelAgent', 'LoopAgent']
random.shuffle(workflow_types)
MY_WORKFLOW = workflow_types[0]

print(f'✅ English!')
print(f'   📊 English: {[t["topic"] for t in my_topics]}')
print(f'   📦 Chunks: {len(my_chunks)} | Collection: {collection_name}')
print(f'   🔍 Query English: "{MY_QUERY}"')
print(f'   🔧 Tool English: {MY_TOOL["name"]}')
print(f'   🔄 Workflow English: {MY_WORKFLOW}')
```

## 🔧 English (English)

```python
# ===== English cell English =====
async def chat_with_agent(agent, message, user_id='user_1', session_id=None):
    """English Agent English"""
    session_service = InMemorySessionService()
    if session_id is None:
        session_id = f'session_{id(agent)}'
    session = await session_service.create_session(
        app_name=agent.name, user_id=user_id, session_id=session_id)
    runner = InMemoryRunner(
        agent=agent, app_name=agent.name, session_service=session_service)
    content = types.Content(role='user', parts=[types.Part.from_text(text=message)])
    
    final_response = ''
    tool_calls = []
    async for event in runner.run_async(
        user_id=user_id, session_id=session.id, new_message=content):
        if event.content and event.content.parts:
            for part in event.content.parts:
                if part.text: final_response = part.text
                if part.function_call:
                    tool_calls.append(part.function_call.name)
    
    if tool_calls:
        print(f'  🔧 Tools: {tool_calls}')
    return final_response, session.id

print('✅ English chat English')
```

```python
# RAG Tool (English — English)
def search_knowledge(query: str) -> dict:
    """English AI English
    
    Args:
        query: English
    """
    q_vec = embed_model.encode(f'query: {query}').tolist()
    results = qdrant.query_points(
        collection_name=collection_name, query=q_vec, limit=3).points
    return {
        'status': 'success',
        'results': [{'text': r.payload['text'], 'source': r.payload['source'],
                     'score': round(r.score, 4)} for r in results]
    }

print('✅ RAG Tool English')
print(f'\n📌 Tool English: {MY_TOOL["name"]}')
print(f'   English: {MY_TOOL["func"]}({MY_TOOL["params"]})')
print(f'   English return dict')
```
