````python
# ─── English English Agent English Colab ───
import asyncio

async def chat_with_agent(runner, message, user_id='user_1', session_id='default_session'):
    """English Agent English"""
    content = types.Content(
        role='user',
        parts=[types.Part.from_text(text=message)]
    )
    
    final_response = ''
    tool_calls = []
    
    async for event in runner.run_async(
        user_id=user_id,
        session_id=session_id,
        new_message=content
    ):
        if event.content and event.content.parts:
            for part in event.content.parts:
                if part.text:
                    final_response = part.text
                if part.function_call:
                    tool_calls.append({
                        'name': part.function_call.name,
                        'args': dict(part.function_call.args) if part.function_call.args else {}
                    })
    
    if tool_calls:
        print(f'  🔧 Tools English: {[t["name"] for t in tool_calls]}')
    
    return final_response


async def demo_chat(agent, messages):
    """Demo English Agent (English session English)"""
    runner = InMemoryRunner(agent=agent, app_name=agent.name)
    session_id = f'session_{id(agent)}'
    
    # English session English runner
    await runner.session_service.create_session(
        app_name=agent.name,
        user_id='user_1',
        session_id=session_id
    )
    
    for msg in messages:
        print(f'\n👤 English: {msg}')
        response = await chat_with_agent(runner, msg, session_id=session_id)
        print(f'🤖 Agent: {response}')


print('✅ English chat English')
````

````python
%%time
# ─── Tool 1: English BMI ───
def calculate_bmi(weight_kg: float, height_cm: float) -> dict:
    """English (BMI) English

    Args:
        weight_kg: English
        height_cm: English

    Returns:
        dict: English BMI, English, English
    """
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)

    if bmi < 18.5:
        level = 'English'
        advice = 'English'
    elif bmi < 25:
        level = 'English'
        advice = 'English English'
    elif bmi < 30:
        level = 'English'
        advice = 'English'
    else:
        level = 'English'
        advice = 'English'

    return {'bmi': round(bmi, 1), 'level': level, 'advice': advice}


# ─── Tool 2: English ───
def convert_temperature(value: float, from_unit: str) -> dict:
    """English Celsius English Fahrenheit

    Args:
        value: English
        from_unit: English ('celsius' English 'fahrenheit')

    Returns:
        dict: English
    """
    if from_unit.lower() == 'celsius':
        result = (value * 9/5) + 32
        return {'result': round(result, 1), 'from': f'{value}°C', 'to': f'{result:.1f}°F'}
    elif from_unit.lower() == 'fahrenheit':
        result = (value - 32) * 5/9
        return {'result': round(result, 1), 'from': f'{value}°F', 'to': f'{result:.1f}°C'}
    else:
        return {'error': f'English {from_unit}'}


# ─── English Agent English Tools ───
health_agent = Agent(
    name='health_agent',
    model='gemini-2.5-flash',
    description='Agent English English BMI English',
    instruction="""English Agent English English
    - English calculate_bmi English BMI English
    - English convert_temperature English
    - English tools English
    """,
    tools=[calculate_bmi, convert_temperature]
)

print('✅ English Health Agent English 2 tools')
print(f'   🔧 Tools: {[t.__name__ for t in health_agent.tools]}')
````

````python
%%time
# ─── English Qdrant VectorDB (English Day 1) ───
import hashlib, random
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient, models

# English Embedding Model
embed_model = SentenceTransformer('intfloat/multilingual-e5-large')
print('✅ English embedding model')

# English Day 1
documents = {
    'healthcare': 'English English AI English X-ray English English 30 English 5 English English Deep Learning English 95% English English English',
    'banking': 'English English KADE AI Assistant English RAG English English 5 English 30 English English English English English',
    'education': 'English English English RAG English-English English PDF English 500 English vector embeddings English multilingual model English English English',
    'agriculture': 'English English AI English English English Computer Vision English English English 92% English English 40%',
    'kmitl': 'English (KMITL) English English AI English Smart Campus English IoT sensor English Machine Learning English English 25% English NLP English English RAG English AI Tutor English',
    'nlp': 'English English NLP English English English Word Segmentation English English PyThaiNLP English English English English English tokenization',
    'rag_architecture': 'English RAG English 3 English English Retriever English VectorDB, Reader English, English Generator English English RAG English hallucination English LLM English',
    'embedding': 'Text Embedding English vector English English multilingual-e5-large English Microsoft English 100 English English vector English 1024 English English vector English embedding',
    'vector_db': 'Qdrant English Vector Database English English ANN (Approximate Nearest Neighbor) English vector English Cosine Similarity, Dot Product, English Euclidean Distance',
}

# Chunk + Embed + Index
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=30)

all_chunks = []
all_sources = []
for source, text in documents.items():
    chunks = splitter.split_text(text)
    for chunk in chunks:
        all_chunks.append(chunk)
        all_sources.append(source)

print(f'📊 {len(all_chunks)} chunks English {len(documents)} English')

# Embed
passages = [f'passage: {c}' for c in all_chunks]
embeddings = embed_model.encode(passages, show_progress_bar=True)
VECTOR_SIZE = embeddings.shape[1]
print(f'✅ Embedding: {embeddings.shape}')

# Create Qdrant collection
qdrant = QdrantClient(':memory:')
qdrant.create_collection(
    collection_name='thai_ai_kb',
    vectors_config=models.VectorParams(size=VECTOR_SIZE, distance=models.Distance.COSINE)
)

# Upsert
points = [
    models.PointStruct(
        id=i,
        vector=embeddings[i].tolist(),
        payload={'text': all_chunks[i], 'source': all_sources[i]}
    )
    for i in range(len(all_chunks))
]
qdrant.upsert(collection_name='thai_ai_kb', points=points)
print(f'✅ Upsert {len(points)} points English Qdrant')
````

````python
%%time
# ─── English RAG Tool ───
def search_thai_ai_knowledge(query: str) -> dict:
    """English AI English English

    English tool English:
    - English AI English (English English English English)
    - English RAG, Embedding, Vector Database
    - NLP English

    Args:
        query: English English

    Returns:
        dict: English
    """
    query_vector = embed_model.encode(f'query: {query}').tolist()
    results = qdrant.query_points(
        collection_name='thai_ai_kb',
        query=query_vector,
        limit=3
    ).points

    if not results:
        return {'status': 'no_results', 'message': 'English'}

    contexts = []
    for r in results:
        contexts.append({
            'text': r.payload['text'],
            'source': r.payload['source'],
            'score': round(r.score, 4)
        })

    return {
        'status': 'success',
        'num_results': len(contexts),
        'results': contexts
    }

print('✅ RAG Tool English')

# English tool English
test = search_thai_ai_knowledge('AI English')
print(f'\n🧪 English: {test["num_results"]} English')
for r in test['results']:
    print(f'  [{r["source"]}] score={r["score"]} | {r["text"][:60]}...')
````

````python
%%time
# ─── English RAG Agent ───
rag_agent = Agent(
    name='thai_ai_expert',
    model='gemini-2.5-flash',
    description='English AI English English',
    instruction="""English AI English

    English:
    1. English AI English, case study, English RAG
       → English search_thai_ai_knowledge English
    2. English English
    3. English English English "English"
    4. English English English → English
    5. English English English bullet points
    """,
    tools=[search_thai_ai_knowledge]
)

print('✅ English RAG Agent English!')
````

````python
%%time
from google.adk.agents import LoopAgent

# Agent 1: English
writer_agent = Agent(
    name='writer_agent',
    model='gemini-2.5-flash',
    description='English',
    instruction="""English English English
    English feedback English English""",
    tools=[search_thai_ai_knowledge]
)

# Agent 2: English (critic)
critic_agent = Agent(
    name='critic_agent',
    model='gemini-2.5-flash',
    description='English English feedback',
    instruction="""English:
    - English?
    - English?
    - English?
    
    English → English
    English → English 'APPROVED: [English]'
    """
)

# LoopAgent: English → English → English → English → ... (English 3 English)
loop_refiner = LoopAgent(
    name='refine_loop',
    description='English',
    sub_agents=[writer_agent, critic_agent],
    max_iterations=3  # English!
)

print('✅ English LoopAgent: English ↔ English (English 3 English)')
print(f'   Max iterations: {loop_refiner.max_iterations}')
````

````python
# ─── Demo: Agent English ───
async def conversation_with_memory(agent, messages):
    """English English session English → Agent English"""
    runner = InMemoryRunner(agent=agent, app_name=agent.name)
    
    await runner.session_service.create_session(
        app_name=agent.name,
        user_id='student_demo',
        session_id='memory_demo'
    )

    for msg in messages:
        print(f'\n👤 English: {msg}')
        content = types.Content(
            role='user',
            parts=[types.Part.from_text(text=msg)]
        )

        final_response = ''
        async for event in runner.run_async(
            user_id='student_demo',
            session_id='memory_demo',
            new_message=content
        ):
            if event.content and event.content.parts:
                for part in event.content.parts:
                    if part.text:
                        final_response = part.text
        
        print(f'🤖 Agent: {final_response}')

# English
await conversation_with_memory(greeting_agent, [
    'English English',
    'English AI English',
    'English',
    'English',
    'English?',
])
````
